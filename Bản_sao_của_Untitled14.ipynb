```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Access stock price of AAPL from 2020 to 2024
df = yf.download('AAPL', start='2020-01-01', end='2024-12-31')
#display first 5 rows
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Close'], label =("Close Price"))
plt.title('AAPL Stock Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()

```

```python
df["50_MA"] = df["Close"].rolling(window=50).mean()
df["200_MA"] = df["Close"].rolling(window=200).mean()

plt.figure(figsize=(12,6))
plt.plot(df["Close"], label="Close Price", alpha=0.6)
plt.plot(df["50_MA"], label="50-Day MA", linestyle="dashed")
plt.plot(df["200_MA"], label="200-Day MA", linestyle="dotted")
plt.title("Xu hng gi vi ng trung bnh ng")
plt.xlabel("Ngy")
plt.ylabel("Gi (USD)")
plt.legend()
plt.show()
```
